```python
# Re-shaping the Dependent feature
y = y.reshape(len(y), 1) # changing hte shape from (50,) to (50,1)
```

#### Feature Scaling
#### Since, Gradient Descent is scale sensiitve, differnt scale and lead to failure. Numerical stability can help in reducing overflow/underlfow and can be acheived by feature scaling.

```python
# Feature Scaling for Independent variable
for i in range(X.shape[1]-2):
    X[:, i] = (X[:, i] - np.mean(X[:, i])) / np.std(X[:, i])
```

```python
# Feature Scaling for Dependent variable
y = (y-int(np.mean(y)))/np.std(y)
```

```python
# Adding X0=1, so we have the equation: y=(W1*X1) + (W0*X0)
X = np.concatenate((X, np.ones((50,1))), axis=1)

# Basically, we are adding a couln of 1s to our matrix of X(independent features)
# It is done, so that Bias (intercept) term is added to the model wihtout writing it seperately
```

#### Train-test Split of Dataset

```python
def split_data(X,y,test_size=0.2, random_state=0):
    np.random.seed(random_state) #seed for reproducible results, so that shuffling doesnot give differnet splits
    indices = np.random.permutation(len(X)) # randomly shuffles row numbers
    data_test_size = int(X.shape[0]*test_size)  # gives the number of samples to add to the test_size
    
    #Seperating the Independent and Dependent features into Train and Tesr set
    train_indices = indices[data_test_size:]  #adding first indices to test_size
    test_indices = indices[:data_test_size]   #adding the remaining to the train_set
    
    #Seperating input and output for training anf testing
    X_train = X[train_indices]
    y_train = y[train_indices]
    
    
    X_test = X[test_indices]
    y_test = y[test_indices]
    
    return X_train, y_train, X_test, y_test
```

#### Traning the Linear Regression Model

```python
class multipleLinearRegression():
    def __init__(self):
        pass  # it sets up the model, no initialization is done here

    def forward(self, X, W):
        return np.dot(X, W)  # y_pred

    def compute_loss(self, y_true, y_pred):
        return np.mean((y_pred - y_true) ** 2) / 2  # predicts profits by multiplying inputs(X) and weights(W)


    def update_weights(self, X, y, W, alpha):
        m = X.shape[0]
        for i in range(m):
            y_pred = np.dot(X[i], W) 
            gradient = (y_pred - y[i]) * X[i] 
            W -= alpha * gradient.reshape(-1, 1)
        return W

    def train(self, X, y, epochs=10, alpha=0.001, random_state=0):
        np.random.seed(random_state)
        m, n = X.shape
        W = np.random.randn(n, 1) / np.sqrt(m) # randomly initializing weights
        y = y.reshape(-1, 1)  # it ensures that y is a column

        train_loss = []   # it wil track error over time
        for epoch in range(epochs):
            indices = np.random.permutation(m)
            X_shuffled = X[indices]
            y_shuffled = y[indices]

            W = self.update_weights(X_shuffled, y_shuffled, W, alpha)
            loss = self.compute_loss(y_shuffled, np.dot(X_shuffled, W))
            train_loss.append(loss) # saves the error for plotting later

        return W, train_loss  

    def test(self, X_test, y_test, W):
        y_pred = np.dot(X_test, W)
        loss = self.compute_loss(y_test.reshape(-1, 1), y_pred)
        return y_pred, loss

    def predict(self, X_sample, W):
        return np.dot(X_sample, W)

    def plot_loss(self, loss):
        plt.plot(range(len(loss)), loss)
        plt.xlabel("Epochs")
        plt.ylabel("Loss")
        plt.title("Training Loss Over Epochs")
        plt.show()
```

```python
# Splitting the dataset (train 80% and test 20%)
X_train, y_train, X_test, y_test = split_data(X,y)
```

```python
regressor = multipleLinearRegression()
```

```python
#training the model for 200 cycles 
num_epochs = 200
W_trained, train_loss = regressor.train(X_train, y_train, epochs=num_epochs, alpha=0.0001)
```

```python
# testing on the test dataset
test_pred, test_loss = regressor.test(X_test, y_test, W_trained)
```

```python
plt.plot(range(1, num_epochs + 1), train_loss, label='Training Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('Training Loss over Epochs')
plt.legend()
plt.show()
```

#### The plot indicates a steady decrease in th erroe value over 200 epochs. The smooth curve confirms that the training was done with a constant learning rate appropriate to the dataset. 
#### The loss decreases steadily over 200 epochs and the model is learning predict profits more accurately. This means, Gradient Descent algorithm effectively adjusts the weights to minimize errors.
